### Пропуски значений

Посчитаем долю пропущенных значений в каждой колонке:

Для начала давайте посмотрим, связано ли отстутсвие данных в столбце `kitchen_area` с тем, что квартира может являться студией без кухни, для этого посмотрим на строки таблицы, где отсутствуют эти значения:

Видимо, в данном столбце отстутсвуют значения по другой причине, давайте так же изучим пропуски в столбце `floors_total`, возможно они указывают на то, что в доме просто один этаж, для этого выведем первые десять значений с пропусками:

```python
df[df['floors_total'].isna()].head(10)
```

Видимо, количество этажей в здании просто не было указано в объявлениях, как и такие же значения в столбцах `ponds_nearest`, `parks_nearest`, `city_centers_nearest`, `airports_nearest`, `kitchen_area`, `living_area` и `ceiling_height`. А в столбце `days_exposition` значения скорее всего пропали по другой причине, так как это программное значение, возможно в некоторых случаях, к примеру, когда квартира очень быстро продалась, иногда происходят сбои и значение не записывается вовсе. С остальными пропусками мы можем поработать.

Давайте заменим пропуски в столбце `is_apartment` на `False`, а в столбце `balcony` на `0`, так как в данном столбце отстуствие значения скорее всего указывает на отстуствие данного элемента в квартире, а в столбце `locality_name` можно пропуски заполнить значением `Неизвестно`, ведь город неизвестен:

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
df['balcony'] = df['balcony'].fillna(0)
df['locality_name'] = df['locality_name'].fillna('Неизвестно')
```

Также мы можем заполнить `floors_total`, `ceiling_height`, `days_exposition`, медианным значением, а также нам понадобятся для исследования значения `city_centers_nearest` для Санкт-Петербурга, это не должно никак повлиять на точность нашего исследования. Остальные поля оставим так, как есть, потому что в них пропусков очень много и если мы их заполним средними значениями, то это может сильно повлиять на результаты нашего исследвания, поэтому в данной ситуации пропуски тоже являются значениями:

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
df['days_exposition'] = df['days_exposition'].fillna(df['days_exposition'].median())
df['floors_total'] = df['floors_total'].fillna(df['floors_total'].median())
df.loc[df['locality_name'] == 'Санкт-Петербург', 'city_centers_nearest'] = (
    df.loc[df['locality_name'] == 'Санкт-Петербург', 'city_centers_nearest'].fillna(
        df.loc[df['locality_name'] == 'Санкт-Петербург', 'city_centers_nearest'].median()
    )
)
```

При заполнении пропусков в колонке `floors_total` могла нарушиться логика с столбцом `floor`, то-есть текущий этаж квартиры будет выше нашей медианы, давайте посчитаем количество таких объявлений:

```python
len(df[df['floor'] > df['floors_total']])
```

Не так уж и много, давайте заполним пропуски более грубо, квартиры на последнем этаже встречаются куда реже, чем квартиры посередине, поэтому давайте прибавим просто один этаж от текущего, таким образом квартира не окажется на последнем этаже и в объявлениях не будет логической ошибки:

```python
df.loc[df['floor'] > df['floors_total'], 'floors_total'] = df['floor'] + 1
```

Проверим наши изменения:

```python
len(df[df['floor'] > df['floors_total']])
```

Чтобы заполнить значения в столбцах `living_area` и `kitchen_area` мы можем рассчитать долю от общей площади в среднем, для каждого столбца:

```python
living_area_percentage = (df['living_area'] / df['total_area']).mean()
kitchen_area_percentage = (df['kitchen_area'] / df['total_area']).mean()
print(f'Доля для living_area: {living_area_percentage}')
print(f'Доля для kitchen_area: {kitchen_area_percentage}')      
```

Теперь давайте умножим общуюю площадь на наши доли, чтобы заполнить пропуски:

```python
df.loc[df['living_area'].isna(), 'living_area'] = df['total_area'] * living_area_percentage
df.loc[df['kitchen_area'].isna(), 'kitchen_area'] = df['total_area'] * kitchen_area_percentage
```

## Расчёты и добавление результатов в таблицу

Добавим цену квадратного метра для каждой квартиры, для этого поделим стоимость на количество квадратный метров:

```python
df['price_square'] = df['last_price'] / df['total_area']
```

Давайте добавим также добавим день недели, месяц и год публикации объявлений:

```python
df['day_of_week'] = pd.DatetimeIndex(df['first_day_exposition']).dayofweek
df['month'] = pd.DatetimeIndex(df['first_day_exposition']).month
df['year'] = pd.DatetimeIndex(df['first_day_exposition']).year
```

Добавим категоризацию этажей для каждой квартиры, будем делить квартиры на три категории:
- Первый этаж
- Другой этаж
- Последний этаж

Также к каждой категории добавим соответсвующий `id` для дальнейшего использования в графиках. Для этого давайте напишем функцию:

```python
FIRST_FLOOR = 'Первый этаж'
OTHER_FLOOR = 'Другой этаж'
LAST_FLOOR = 'Последний'


def get_category_floor(row):
    """ Функция возвращает одну из трёх категорий в
        зависимости от этажа, на вход функция
        принимает строку.
    """
    try:
        # получим из строчки этаж и количество этажей в здании
        floor, floors_total = row['floor'], int(row['floors_total'])
    except ValueError:
        # если количество этажей неизвестно, произойдёт ошибка,
        # вернём в данном случае категорию другой этаж, так как
        # количество этажей в здании неизвестно
        return OTHER_FLOOR
    return (
        FIRST_FLOOR if floor == 1
        else LAST_FLOOR if floor == floors_total
        else OTHER_FLOOR
    )


def get_category_floor_id(category_floor):
    """ Функция возвращает id каждой категории,
        на вход принимает саму категорию этажа.
    """
    return (
        1 if category_floor == FIRST_FLOOR
        else 2 if category_floor == OTHER_FLOOR
        else 3
    )
```

Применим нашу функцию к каждой строчке, результат запишем в новый столбец:

```python
df['category_floor'] = df.apply(get_category_floor, axis=1)
```

Добавим для каждой категории `id`:

```python
df['category_floor_id'] = df['category_floor'].apply(get_category_floor_id)
```

Теперь рассчитаем отношение жилой площади к общей площади:

```python
df['living_to_total'] = df['living_area'] / df['total_area']
```

### Самые популярные, дорогие и дешёвые регионы

Чтобы найти самые популярные, дорогие и дешёвые регионы давайте создадим новую таблицу, в которой рассчитаем количество объявлений для каждого региона, а также среднюю цену за квадратный метр:

```python
# создадим таблицу в которую сгруппируем данные по названию населённого пункта,
# посчитаем количество объявлений и рассчитаем среднее по столбцу "price_squere"
regions = df_filtered.pivot_table(
    index='locality_name',
    values='price_square',
    aggfunc=['count', 'mean'],
)
# переименуем колонки
regions.columns = ['count', 'mean']
# отсортируем по убыванию, по столбцу "count", таким образом получим самые
# популярные регионы
regions.sort_values(by='count', ascending=False).head(10)
```

Чтобы получить самые дорогие и дешёвые населённые пункты, нужно отсеять регионы, где объявлений меньше 10 штук, так как в таких регионах среднее может быть сильно завышено или занижено:

```python
regions = regions[regions['count'] > 10]
```

Теперь отсортируем полученную таблицу по средним значениям, сначала по возрастанию, так мы получим самые дорогие регионы, потом по убыванию:

```python
display(regions.sort_values(by='mean', ascending=False).head(10))
display(regions.sort_values(by='mean').head(10))
```

- посёлок Шушары

Самые дорогие населённые пункты:

- Санкт-Петербург — стоимость квадратного метра тут больше 110 тыс. рублей
- Пушкин — стоимость квадратного метра больше 100 тыс. рублей

## Общий вывод

В основном квартиры продаются от 3 млн до 6 млн рублей, с потолками размером от 2.4 до 3 метров, площадью от 25 до 100 квадратных метров и количестом комнат до трёх. Время продажи может сильно отличаться, можно считать, что квартира продалась быстро, если продажа была совершенна до трёх месяцев от публикации объявления.

На стоимость квартиры больше всего влияет площадь, после количество комнат, что не удивительно, ведь скорее всего от количества комнат также зависит и площадь, а также сильно влияет на цену этаж, если квартира на первом этаже, скорее всего она будет более дешёвая, чем на любом другом этаже, на последнем этаже чуть больше дорогих квартир, чем на первом, в основном все дорогие квартиры расположенны на других этажах, потом влияет но не так сильно расположение квартиры по отношению к центру города, самое маленькое влияние было замечено в зависимости от даты публикации, единственно, что можно сказать про дату публикации: год от года цены на квартиры меняются, иногда они возрастают, иногда падают.

Большее количество объявлений оказалось в Санкт-Петербурге, и на удивление в посёлках Мурино и Шушары, самые дорогие квартиры в Зеленограде и Санкт-Петербурге, а самые дешёвые в Бокситогорске и Сланцах.

В Санкт-Петербурге достаточно сильно заметна зависимость цены квартиры от расстояния до центра города, в пределах восьми километров встречаются самые дорогие квартиры во всём Питере, дальше цены резко падают. Если сравнивать центр со всем городом, то видны некоторые отличия: в центре квартиры в основном от 50 до 100 квадратных метров, когда во всём городе распределение имеет более острый пик от 40 до 60 метров, дешёвых предложений в центре мало, во основном квартиры в центре стоят дорого, особо дорогие предложения встречаются в основном только в центре. В центре мало однушек, в основном двушки и трёшки, когда во всём городе больше однушек и двушек, где-то около половины четырёхкомнатных квартир приходится на центр, редкие квартиры с большим количетсвом комнат в основном все из центра. Также только в центре встречаются квартиры с высокими потолками.

В центре чем больше комнат, тем реже встречаются дешёвые квартиры, больше всего дорогих квартир среди трёх и четырёхкомнтных квартир. В центре этаж сильнее влияет на стоимость квартиры, чем во всём городе, а удалённость от центра наоборот играет меньшую роль, чем во всём городе.
